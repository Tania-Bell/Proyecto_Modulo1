## Definición de hipótesis y supuestos

>Los supuestos que se dan por verdadero en nuestro proyecto son los siguientes:

## Obtención de bases de datos

> Los datos se obtuvieron de una plataforma de diferentes bases de datos llamada Kaggle. La base de datos seleccionada es de la plataforma de YouTube, con información de EEUU hasta el año 2019.

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = pd.read_csv('USvideos.csv').dropna()
data = data[['views','likes','dislikes','comment_count']]
data = data[data['views']>60000000]
data = data[data['likes']>600000]
data = data[data['dislikes']<600000]
data = data[data['comment_count']>6000]
data
```

```python
corrMatrix = data.corr()
sn.heatmap(corrMatrix, annot=True, linewidths=.0001)
sn.set(rc = {'figure.figsize':(10,8)})
plt.show()
```
