# Proyecto Simulación - Predicción de prima anual de seguro médico

## Descripción

> En este proyecto utilizaremos los datos reales de una empresa de seguros médicos. Con datos de casi 1000 clientes, crearemos un modelo que prediga cuanto será la prima anual que pagarán los futuros clientes. Los datos incluyen parámetros entorno a la salud de los clientes y se determinará cuales datos son los más influyentes en el cálculo de la prima. Los datos incluyen la prima del seguro que pagó cada cliente. 

## Objetivos

> #### Objetivo General:
El objetivo de este proyecto es obtener la prima anual que paga el cliente para obtener seguro médico.
> #### Objetivo específico:
    > - Determinar la prima anual que paga el cliente, basándose en parámetros de su salud.
    > - Determinar cuáles son los factores de salud de los clientes que varian el costo de la prima.
    > - Incluso se puede analizar la influencia que tiene el peso y la edad del cliente en el costo de la prima.

## Definición del Problema

>Las empresas de seguros necesitan un sistema para poder determinar la prima de seguro de sus clientes. Esta prima tiene que ser personalizada tomando en cuenta las diferentes condiciones de salud de cada cliente. Ningún cliente va a tener el mismo estado de salud ni histórico que otro, y es por eso que se busca un modelo que se adapte a las diferentes condiciones. El seguro necesita tener las herramientas para realizar el cálculo de cuánto le va a costar cubrir los gastos médicos de dicha persona y hacer el cálculo del riesgo que tiene el cliente de tener un problema de salud severo y por lo tanto costoso. Los seguros ayudan a las personas pero no son organizaciones, son negocios y necesitan tener una rentabilidad. Nosotros queremos encontrar un modelo que determine esa prima que paguen los clientes, tomando en cuenta su situación de salud.

## Nodos y variables que se decidieron simular y por qué

>Para desarrollar el proyecto tomamos las siguientes variables:
    > - Edad
    > - Diabetes
    > - Presión arterial
    > - Transplante de órganos
    > - Enfermedades crónicas
    > - Altura
    > - Peso
    > - Alergias
    > - Antecedentes de cáncer en la familia
    > - Número de cirugías
    
>Estas variables se eligeron porque son unas de las más importantes cuando se calcula la prima ya que dan un buen panorama de la salud del cliente y el riesgo que tiene de tener problemas de salud en el futuro.

## Definición de hipótesis y supuestos

>Los supuestos que se dan por verdadero en nuestro proyecto son los siguientes:
    > - Los 1000 clientes representan todos los clientes de la aseguradora.
    > - El modelo para obtener cálculo y análisis de los diferentes clientes es el mismo.
    > - Las variables que se analizarán son suficientes para determinar la prima del seguro.

>Nuestra hipótesis es que el factor que más influye en el precio de la prima para el seguro médico es la edad. Llegamos a esta conclusión porque entre más grande es una persona, más problemas de la salud tendrá. Las personas jóvenes pueden vivir una vida no muy sana pero no tener ningun problema de la salud, pero cuando van creciendo más problemas se van desarrollando ya que el cuerpo se va desgastando.
